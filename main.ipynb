```python
import pandas as pd
from preprocess import token

#import dataset
train = pd.read_csv("data/train.csv",encoding = "ISO-8859-1")
test = pd.read_csv("data/test.csv",encoding = "ISO-8859-1")

#import the processed text
train_processed= pd.read_csv("data/train_processed.csv", encoding = "ISO-8859-1")
test_processed= pd.read_csv("data/test_processed.csv",encoding = "ISO-8859-1")

#concatenate the processed files and the initial dataset
train = pd.concat([train, train_processed], axis = 1,join_axes=[train.index])
train = train.drop("Unnamed: 0", axis = 1)
test = pd.concat([test, test_processed], axis = 1,join_axes=[test.index])
test = test.drop("Unnamed: 0", axis = 1)
```

```python
from nltk import ngrams
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
train.fillna(' ',inplace=True)
test.fillna(' ',inplace=True)
vectWord = TfidfVectorizer(lowercase=True, analyzer='word',ngram_range=(1,2),dtype=np.float32)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
def train_classifier(X, y):
    """Train a classifier using the given training data.
    Trains a logistic regression on the input data with default parameters.
    """
    cls = LogisticRegression(C=10)
    cls.fit(X, y) 
    return cls

def evaluate(X, yt, cls):
    """Evaluated a classifier on the given labeled data using accuracy."""
    yp = cls.predict(X)
    acc = metrics.accuracy_score(yt, yp)
    print("  Accuracy", acc)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(train.Processed, train.label, test_size=0.4, random_state=1)
trainVector = vectWord.fit_transform(X_train)
validVector = vectWord.transform(X_valid)
```

```python
cls = train_classifier(trainVector, y_train)
evaluate(validVector, y_valid, cls)
```
